## 14 Converter

```python
def name_abbrevation (name):
    """Create Abbrevation For any given name"""
    if ' ' in name:
        strings = name.split(" ")
        # print(strings)
        abbre = [alphabets[0]  for alphabets in strings]
        
        abbre_final = ""
        for i in abbre:
            abbre_final = abbre_final+i
        return abbre_final
    else:
        return name
```

```python
pd.read_csv(path+'IPL Matches 2008-2020.csv', converters={'team1': name_abbrevation, 'team2':name_abbrevation}, nrows=10)
```
